```python
#Predict survival rate of passenger
import pandas as pd
```

```python
df = pd.read_csv('H:/Python In Data Science/titanic.csv')
df
```

```python
df['Age'].fillna(0, inplace=True)
df.head()
```

```python
t_df = df.drop(['PassengerId','Name'],axis='columns')
t_df.head()
```

```python
#inputs = df.drop('Survived',axis='columns') -- either drop Target/independent column  or
inputs = t_df[['Pclass','Sex','Age','Fare']]
target = t_df['Survived']
```

```python
# convert 'Sex' col. using LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder() # object created 
```

```python
inputs['Sex_n'] = le_sex.fit_transform(inputs['Sex']) # fit_transform - Encoded successfully 
inputs.head()
```

```python
inputs_n = inputs.drop('Sex',axis='columns') # drop orignal col.which is no use
inputs_n.head()
```

```python
#from sklearn.model_selection import train_test_split # 
```

```python
# we have I/P and O/p arguments from trainig stets and we divided into test and train sets
#X_train, X_test, y_train, y_test = train_test_split(inputs_n,target,test_size=0.2)
```

```python
from sklearn import tree # import tree
```

```python
t_model = tree.DecisionTreeClassifier()
```

```python
t_model.fit(inputs_n,target)
```

```python
t_model.score(inputs_n,target) # Model accuracy is 98% approx.
```
